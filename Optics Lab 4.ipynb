```python
import numpy as np
import matplotlib.pyplot as plt
import math
from math import asin
from math import acos
from cmath import sqrt
```

$$r_{\perp} = r_{TE} = \frac{E_r}{E_i} = \frac{n_icos\theta_{i}-n_tcos\theta_t}{n_icos\theta_i+n_icos\theta_t}$$

$$r_{\parallel} = r_{TM} = \frac{E_r}{E_i} = \frac{n_tcos\theta_{i}-n_icos\theta_t}{n_tcos\theta_i+n_icos\theta_t}$$

$$
R_{\parallel} = r_{\parallel}r^*_{\parallel}
$$

$$
R_{\perp} = r_{\perp}r^*_{\perp}
$$

$$
n_isin\theta_i = n_tsin\theta_t
$$

$$
sin\theta_t = \frac{n_i}{n_t}sin\theta_i
$$

```python
theta = np.linspace(0,90,90)

theta_i = np.radians(theta)

def transmittance(n_i,n_t,theta_i):
    theta_t = []
    for i in range(len(theta_i)):
        transmittance_sin = (n_i)/(n_t)*np.sin(theta_i[i])
        print(transmittance_sin)
        if transmittance_sin < 1 and transmittance_sin > -1:       
            transmittance = np.radians(asin((n_i)/(n_t)*np.sin(theta_i[i])))
            theta_t.append(transmittance)
        else:
            return theta_t,i
            break
    return theta_t,i

def r_perp(n_i,n_t,theta_i):
    r_perp_arr = []
    theta_t = transmittance(n_i,n_t,theta_i)[0]
    max_index = transmittance(n_i,n_t,theta_i)[1]
    for i in range(max_index):
        r_perp_num = n_i*np.cos(theta_i[i])-n_t*np.cos(theta_t[i])
        r_perp_den = n_i*np.cos(theta_i[i])+n_i*np.cos(theta_t[i])
        r_perp = r_perp_num/r_perp_den
        r_perp_arr.append(r_perp)
    return r_perp_arr,i

def r_parallel(n_i,n_t,theta_i):
    r_parallel_arr = []
    theta_t = transmittance(n_i,n_t,theta_i)[0]
    max_index = transmittance(n_i,n_t,theta_i)[1]
    for i in range(max_index):
        r_parallel_numer = n_t*np.cos(theta_i[i])-n_i*np.cos(theta_t[i])
        r_parallel_den = n_t*np.cos(theta_i[i])+n_t*np.cos(theta_t[i])
        r_parallel = r_parallel_numer/r_parallel_den
        r_parallel_arr.append(r_parallel)
    return r_parallel_arr,i

```

```python
n_i = 1
n_t = 1.5
r_te = r_perp(n_i,n_t,theta_i)
r_tm = r_parallel(n_i,n_t,theta_i)
```

```python
index_e = r_te[1]
index_m = r_tm[1]
if index_e < len(theta_i)-2:
    text = 'The brewester angle is a little beyond' + str((math.degrees(theta_i[index_e+1])))
else:
    text = 'There is no total internal reflection'
plt.plot(theta_i[0:index_e+1],r_te[0],color='m',label = '$r_{TE}$')
plt.plot(theta_i[0:index_m+1],r_tm[0],color='y',label = '$r_{TM}$')
plt.text(0,0,text)
plt.title('$r_{TE}$ and $r_{TM}$ as a function of $\\theta_{i}$')
plt.xlabel('Angle [rad]')
plt.ylabel('Transmittance')
plt.legend()
plt.show()

print(index_e,index_m,len(theta_t))
```
